```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

### Director's analysis

```python
df = pd.read_csv('netflix.csv')
```

```python
df['release_year'].min()
```

```python
df.replace(['nan', np.nan], inplace = True)
```

```python
df['director'] = df['director'].apply(lambda x: str(x).split(', '))
```

```python
df = df.explode('director')
```

```python
df['director'].value_counts(dropna = True).head()
```

### Top director's movies/shows over the years

### No. of movies and tv series added to netflix over the years

```python
df = pd.read_csv('netflix.csv')
```

```python
df.replace(['nan', np.nan], inplace = True)
```

```python
def year(x):
    return str(x).split()[2][2:4]
        
```

```python
sns.histplot(data = df, x = df['date_added'].apply(year), hue = 'type')
plt.show()
```

### Top Countries who are producing movies/tv shows over the years

```python
df = pd.read_csv('netflix.csv')
```

```python
df.replace(['nan', np.nan], inplace = True)
```

```python
df['country'] = df['country'].apply(lambda x: str(x).split(', '))
df = df.explode('country')
```

```python
df['country'].value_counts().head(5)
```

```python
df.reset_index(inplace = True)
```

```python
sns.histplot(data = df[(df['country'] == 'United States')|(df['country'] == 'India')|(df['country'] == 'United Kingdom')], 
             x = 'release_year', 
             hue = 'country')
plt.xticks(np.arange(2000, 2022, 2))
plt.xlim((2000, 2022))
plt.show()
```
